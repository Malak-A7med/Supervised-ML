```python
# Encodeing Data 
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Gender'] = encoder.fit_transform(df['Gender'])
df['Vehicle_Age'] = encoder.fit_transform(df['Vehicle_Age'])
df['Vehicle_Damage'] = encoder.fit_transform(df['Vehicle_Damage'])
```

```python
df
```

```python
df.corr()['Response'].sort_values()
```

```python
df.drop(['id', 'Region_Code'], axis=1, inplace=True)
```

```python
features = df.drop("Response", axis=1)
target = df['Response']
```

```python
# Check for Unbalance data
target.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
over_sampling = RandomOverSampler()
under_sampling = RandomUnderSampler()
smote = SMOTE()
```

```python
featuer_over, tagret_over = over_sampling.fit_resample(features, target)
featuer_under, tagret_under = under_sampling.fit_resample(features, target)
smote_feat, smote_target = smote.fit_resample(features, target)
```

```python
tagret_over.value_counts()
```

```python
tagret_under.value_counts()
```

```python
smote_target.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, stratify=target, random_state=42)
x_train_over, x_test_over, y_train_over, y_test_over = train_test_split(featuer_over, tagret_over, test_size=0.2, stratify=tagret_over, random_state=42)
x_train_under, x_test_under, y_train_under, y_test_under = train_test_split(featuer_under, tagret_under, test_size=0.2, stratify=tagret_under, random_state=42)
x_train_smote, x_test_smote, y_train_smote, y_test_smote = train_test_split(smote_feat, smote_target, test_size=0.2, stratify=smote_target, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
sfc = RandomForestClassifier(n_estimators=10)
sfc.fit(x_train, y_train)
```

```python
sfc_over = RandomForestClassifier(n_estimators=10)
sfc_over.fit(x_train_over, y_train_over)
```

```python
sfc_under = RandomForestClassifier(n_estimators=10)
sfc_under.fit(x_train_under, y_train_under)
```

```python
sfc_smote = RandomForestClassifier(n_estimators=10)
sfc_smote.fit(x_train_smote, y_train_smote)
```

```python
y_pred = sfc.predict(x_test)
y_pred_over = sfc_over.predict(x_test_over)
y_pred_under = sfc_under.predict(x_test_under)
y_pred_smote = sfc_smote.predict(x_test_smote)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(f"Accuracy of original data: {accuracy_score(y_pred, y_test)}")
print(f"Accuracy of Over sampling data: {accuracy_score(y_pred_over, y_test_over)}")
print(f"Accuracy of Under sampling data: {accuracy_score(y_pred_under, y_test_under)}")
print(f"Accuracy of SMOTE data: {accuracy_score(y_pred_smote, y_test_smote)}")
```

```python
print(f"Accuracy of original data: \n{classification_report(y_pred, y_test)}")
print(f"Accuracy of Over sampling data: \n{classification_report(y_pred_over, y_test_over)}")
print(f"Accuracy of Under sampling data: \n{classification_report(y_pred_under, y_test_under)}")
print(f"Accuracy of SMOTE data: \n{classification_report(y_pred_smote, y_test_smote)}")
```
