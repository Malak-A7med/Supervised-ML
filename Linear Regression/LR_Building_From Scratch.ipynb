```python
import numpy as np
import matplotlib.pyplot as plt

# Training data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# initialize parameters
w = 0
b = 0

# Learning rate and number of epochs
alpha = 0.01
epochs = 1000

# size of training data
m = len(x)
```

```python
costs = []  # to draw the cost over epochs

for i in range(epochs):
    y_pred = w * x + b
    error = y_pred - y

    # calc Derivatives for gradient descent
    dw = (1/m) * np.dot(error, x)
    db = (1/m) * np.sum(error)

    # update parameters
    w = w - alpha * dw
    b = b - alpha * db

    # calc cost for monitoring
    cost = (1/(2*m)) * np.sum(error ** 2)
    costs.append(cost)
```

```python
plt.plot(costs)
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function Decrease Over Time")
plt.show()
```

```python
plt.scatter(x, y, color='blue', label='Real Data')
plt.plot(x, w*x + b, color='red', label='Predicted Line')
plt.legend()
plt.show()

print(f"Slope (w): {w}")
print(f"Intercept (b): {b}")
```
